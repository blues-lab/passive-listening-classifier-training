```python
from transformers import pipeline
```

```python
text_generation = pipeline("text-generation")
```

```python
seed = "The gyms just opened up here in LA"
for i in range(30):
    random.seed(i)
    generated_text = text_generation(seed, max_length=150, do_sample=True, clean_up_tokenization_spaces=True)[0]['generated_text']
    print("\n\n")
    print(generated_text)
```
